```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def fermi(E, E_f, k, T):
    f = (1 + np.exp((E-E_f)/(k*T)))**-1
    return f

energy = np.arange(0, 2.01, 0.01)
E_f = 1
k = 8.62*10**-5
T_1 = 293
T_2 = 2000

ffermi1 = [fermi(e, E_f, k, T_1) for e in energy]
ffermi2 = [fermi(e, E_f, k, T_2) for e in energy]

plt.plot(ffermi1, energy, label='T = 293K')
plt.plot(ffermi2, energy, label='T = 2000K')
plt.ylabel("Energy (eV)")
plt.xlabel("Fermi function F(E)")
plt.legend(loc='upper right')
plt.show()
```

```python
def f_prime(x):
    f_prime = (-np.exp(x))/(np.exp(x)+1)**2
    return f_prime
    
x = np.arange(-40, 41, 1)

f = [f_prime(i) for i in x]

plt.plot(x, f)
plt.xlabel('x')
plt.ylabel("f '(x)")
plt.show()
```
